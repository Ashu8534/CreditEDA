### We can create function to plot bargraph

```python
def createbar(x,y,labe='',titl='',xlabe='',ylabe='',size=(10,5)):
    plt.figure(figsize=size)
    plt.bar(x,y,label=labe)
    plt.title(titl,color='purple',size=15,weight='bold')
    plt.xlabel(xlabe,color='purple',size=15,weight='bold')
    plt.ylabel(ylabe,color='purple',size=15,weight='bold')
    plt.grid()
    plt.legend()
    plt.show()
    
```

```python
createbar(x=s.index,y=s.values,labe='Loan',titl='Loan Data',xlabe='Loan Type',ylabe='Count',size=(10,3))
```

### Lets Categories Income total 

#### Create a function 

```python
def income_total(salary):
    if (salary<=50000):
        return '0-50K'
    elif (salary<=100000):
        return '50K-1L'
    elif (salary<=150000):
        return '1L-1.5L'
    elif (salary<=200000):
        return '1.5L-2L'
    elif(salary<=250000):
        return '2L-2.5L'
    else:
        return '3L-Above'
    
income_total(239606)
```

```python
df['income range'] = df['amt_income_total'].apply(income_total)
df.head(4)
```

```python
s = df['income range'].value_counts()
s.values
idf = pd.DataFrame(s)
idf = idf.reset_index()
idf
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=idf,x='income range',y='count',edgecolor='black')
for i,count in enumerate(idf['count']):
    plt.text(x=i,y=count,s=str(count),ha='center',va='bottom')
plt.show()
```

### Income range of amount credit and amount annuity

```python
df['amt_credit'].min()
df['amt_credit'].max()
```

```python

def credit_income_range(credit):
    if(credit<=100000):
        return '0-1L'
    elif(credit<=500000):
        return '1L-5L'
    elif(credit<=1000000):
        return '5L-10L'
    elif(credit<=1500000):
        return '10L-15L'
    elif(credit<=2000000):
        return '15L-20L'
    elif(credit<=2500000):
        return '20L-25L'    
    elif(credit<=3000000):
        return '25L-30L'
    else:
        return '35L-Above'
    
credit_income_range(23400000)
        
```

```python
df['credit_range'] = df['amt_credit'].apply(credit_income_range)
df.head(5)
```

```python
cdf = df['credit_range'].value_counts()
cdf = cdf.reset_index()
cdf
```

### Plotting bargraph 

```python
plt.figure(figsize=(10,4))
sns.barplot(data=cdf,x='credit_range',y='count',edgecolor='black')
plt.show()
```

### Finding average package of each Occupation 

```python
df.head(1)
```

```python
odf = df.groupby('occupation_type').agg(
    avgpack=('amt_income_total','mean')
).sort_values(by='avgpack',ascending=False).reset_index()
odf['avgpack'] = odf['avgpack'].round(2)
odf
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=odf,x='occupation_type',y='avgpack',edgecolor='black',palette='magma',)
plt.xticks(rotation=45)
plt.title('Average package of each Occupation',color='purple',weight='bold',size=12)
plt.show()
```
